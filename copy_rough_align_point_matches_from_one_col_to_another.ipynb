```python
zvalues = range(ex['minz'], ex['maxz']+1)

for z in zvalues:
    matches = renderapi.pointmatch.get_matches_outside_group(ex['source_col'], float(z), render=render)
    print(z, len(matches))
    renderapi.pointmatch.import_matches(ex['target_col'], matches, render=render)
```
